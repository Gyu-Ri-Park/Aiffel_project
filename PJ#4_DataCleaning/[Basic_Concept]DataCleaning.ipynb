```python
def age_func(x):
    if x>80:
        return 80
    else:
        return x
```

```python
#데이터에 한 줄씩 함수를 적용해주기 위해서 apply를 뒤에 사용한다

data['Age'] = data['Age'].apply(age_func)
```

```python
data['Age'].sort_values()
```

```python
def new_age_func(x):
    if x>70:
        return 70
    else:
        return x
```

#### 개념 정리 
- str 메소드: 문자열 데이터에 적용됩니다. 예로 문자열을 대문자로 변환하거나, 문자열을 분할하는 등의 작업이 가능합니다.  
- dt 메소드 : 날짜/시간 데이터에 적용됩니다. 예로 연도, 월, 일을 추출하거나, 주의 날짜를 계산하는 등의 작업이 가능합니다.
