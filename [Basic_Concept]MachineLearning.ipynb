```python
# Q. 다음 df 변수의 데이터에서 '가성비'라는 칼럼을 만들고,
# '호수' / '가격' 의 값이 0.0008을 넘으면 '가성비' 칼럼 값을 'Good' 아니면 'Bad'로 넣어주세요.
# 힌트 : astype(), apply, lambda를 활용합니다.
df = pd.read_csv('new_modudak.csv')
df

df['호수'] = df['호수'].str.replace('호','')
df['호수'] = df['호수'].astype(int)
df.info()

df['가성비'] = df.apply(lambda row: 'Good' if row['호수'] / row['가격'] > 0.0008 else 'Bad', axis=1)
df
```

```python
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
```

# 머신러닝 : 분류

### 검증데이터 분리

```python
from sklearn.model_selection import train_test_split
```

```python
#train데이터 확인(X)
#random_state : split을 잘해서/ model이 좋아서인지 알수없으니까 random_state고정해서 split 을 고정하는 것

X_train, X_test, y_train, y_test = train_test_split(
    cancer_df.drop('target', axis=1),
                    cancer_df['target'],
                    test_size=0.3, random_state = 1004)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#y는 target이 1개기 때문에 이렇게 나옴

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 의사결정 나무 (Decision Tree)

```python
#머신러닝(분류)
# X_train, y_train 으로 학습
# X_test로 예측값 & y_test(실제값) 과 비교 -> 모델의 정확도 확인

from sklearn.tree import DecisionTreeClassifier

#모델 선택
model = DecisionTreeClassifier()

#학습
model.fit(X_train, y_train)

#예측
pred = model.predict(X_test)
pred
```

### 평가 (accracy)

```python
# 정확도 accuracy_score실제값, 예측값)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

- 89% 의 정확도면 괜찮은 성능이다
